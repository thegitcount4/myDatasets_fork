```python
import pandas as pd
import os

# Variation 1: Split by number of rows
def split_csv_by_rows(input_file, rows_per_file, output_prefix='split_file'):
    # Read CSV with explicit quote character
    df = pd.read_csv(
        input_file,
        # quotechar='"',       # Check if there are cases in which it's necessary to change the quote character
        # on_bad_lines='skip'  # Skip bad lines (CAUTION: use just in specific cases with problematic files; i needed it with car_prices.csv)
        )
    
    # Calculate total number of rows
    total_rows = len(df)

    # Convert rows_per_file to int if it's a float (as if i pass it like 2e5... that returns a float)
    rows_per_file = int(rows_per_file)
    
    # Calculate number of output files needed
    num_files = (total_rows // rows_per_file) + (1 if total_rows % rows_per_file else 0)
    
    # Create output directory if it doesn't exist
    os.makedirs('output', exist_ok=True)
    
    # Split and save files
    for i in range(num_files):
        start_idx = i * rows_per_file
        end_idx = min((i + 1) * rows_per_file, total_rows)
        
        # Create chunk
        chunk = df.iloc[start_idx:end_idx]
        
        # Generate output filename
        output_file = f'output/{output_prefix}_part_{i+1}.csv'
        
        # Save to CSV
        chunk.to_csv(
            output_file, 
            index=False,          # Don't write row index
            # compression='gzip',   # Compress output files
            # encoding='utf-8',     # Specify encoding
            # chunksize=10000       # Process in chunks for very large files
            )
        
        print(f'Created {output_file} with {len(chunk)} rows')
 
```

```python
# Example usage
if __name__ == "__main__":
    # Specify your input CSV file
    input_csv = '/home/hernan/Downloads/python/1_Numpy_and_Pandas/project_data/project_transactions.csv'
    
    # Example 1: Split into files with 1000 rows each
    split_csv_by_rows(input_csv, rows_per_file=2e5, output_prefix='project_transactions')
    
```
